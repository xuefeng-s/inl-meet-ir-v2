```python
import nltk
nltk.download('punkt')


def onlyOneSentence(entry):
  liste = nltk.sent_tokenize(str(entry))
  if len(liste)>1:
    return False
  else:
    return True

data = pd.read_excel('/content/gdrive/MyDrive/Praxisprojekt/Data_AllFour0.xlsx', sheet_name = 'sentences')
newData = pd.DataFrame({'Sentence':["This sentence is awesome"], 'SUBJlang':[3], 'SUBJlang01':[1]})

for ind in data.index:
  if onlyOneSentence(data['Sentence'][ind]):
    df = pd.DataFrame({'Sentence':[data['Sentence'][ind]], 'SUBJlang': [data['SUBJlang'][ind]], 'SUBJlang01':[data['SUBJlang01'][ind]]})
    newData = newData.append(df , ignore_index=True)

newData.to_excel('/content/gdrive/MyDrive/Praxisprojekt/output.xlsx')
print("writing output finished")
```

```python
import nltk
import numpy as np
import re

def moreThanTwoWords(sentence):
  word_count = len(re.findall(r'\w+', sentence))
  if word_count < 3:
    return False
  else:
    return True

def kickOutSmallSentences(liste):
  long_liste = list()
  for sentence in liste:
    if moreThanTwoWords(sentence):
      long_liste.append(sentence)
    else:
      print("kicked out ", sentence)
  return long_liste


# txt objective training in dataframe packen
file_o_train = open('/content/gdrive/MyDrive/Praxisprojekt/objective_cv_train.txt')
liste_o_train = file_o_train.readlines()
count = 0
liste_o_train = kickOutSmallSentences(liste_o_train)
data_o_train = pd.DataFrame({'Sentence': liste_o_train, 'SUBJopin01': 0})

for line in liste_o_train:
    count += 1
    print("Line{}: {}".format(count, line.strip()))

# txt objective testing in dataframe packen
file_o_test = open('/content/gdrive/MyDrive/Praxisprojekt/objective_test.txt')
liste_o_test = file_o_test.readlines()
count = 0
liste_o_test = kickOutSmallSentences(liste_o_test)
data_o_test = pd.DataFrame({'Sentence': liste_o_test, 'SUBJopin01': 0})

for line in liste_o_test:
    count += 1
    print("Line{}: {}".format(count, line.strip()))

# txt subjective training in dataframe packen
file_s_train = open('/content/gdrive/MyDrive/Praxisprojekt/subjective_cv_train.txt')
liste_s_train = file_s_train.readlines()
count = 0
liste_s_train = kickOutSmallSentences(liste_s_train)
data_s_train = pd.DataFrame({'Sentence': liste_s_train, 'SUBJopin01': 1})

for line in liste_s_train:
    count += 1
    print("Line{}: {}".format(count, line.strip()))

# txt subjective testing in dataframe packen
file_s_test = open('/content/gdrive/MyDrive/Praxisprojekt/subjective_test.txt')
liste_s_test = file_s_test.readlines()
count = 0
liste_s_test = kickOutSmallSentences(liste_s_test)
data_s_test = pd.DataFrame({'Sentence': liste_s_test, 'SUBJopin01': 1})

for line in liste_s_test:
    count += 1
    print("Line{}: {}".format(count, line.strip()))

# alle dataframes zusammenfügen und zufällig die Sätze mischen
allData = data_o_train.append(data_o_test.append(data_s_train.append(data_s_test)))
allData = allData.sample(frac = 1)

# Dataframe als excel datei schreiben
allData.to_excel('/content/gdrive/MyDrive/Praxisprojekt/opinionDataset.xlsx', index = False)
```
